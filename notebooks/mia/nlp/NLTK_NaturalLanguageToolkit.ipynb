# Stemming

Técnica utilizada para reducir una palabra a su forma raíz o base. Esta forma raíz es la parte central y significativa de una palabra que ignora los sufijos y prefijos. El objetivo del stemming es tratar las diferentes variantes de una palabra como una sola palabra, lo que ayuda a simplificar el análisis de texto y el procesamiento de lenguaje natural.

## Snowball Stemmer

Lenguaje de procesamiento de cadenas para realizar stemming (cálculo de derivaciones de palabras, o dicho de otra manera, calcular la raíz de las palabras o su "stem").

El SnowballStemmer en NLTK es una herramienta que se utiliza para realizar la reducción de palabras (stemming) en un texto. La reducción de palabras, o stemming, es el proceso de eliminar sufijos y prefijos de las palabras para obtener su forma raíz o base, lo que ayuda a reducir las palabras a su forma más simple y común.

El SnowballStemmer en NLTK se basa en el algoritmo Snowball, que es una colección de algoritmos de stemming desarrollados por el lingüista Martin Porter. Estos algoritmos están diseñados para funcionar en varios idiomas y son ampliamente utilizados en tareas de procesamiento de lenguaje natural para normalizar las palabras y reducir la dimensionalidad de los datos de texto.

https://snowballstem.org/

# FreqDist

FreqDist (Frecuencia de Distribución) es una clase en la biblioteca NLTK (Natural Language Toolkit) de Python que se utiliza para calcular y trabajar con la frecuencia de ocurrencia de elementos en una secuencia, como palabras en un texto. La frecuencia de distribución se utiliza comúnmente en tareas de procesamiento de lenguaje natural para analizar la frecuencia de palabras en un corpus de texto y obtener información estadística sobre cuán comunes son ciertas palabras o elementos en un conjunto de datos.

Al crear un objeto FreqDist, puedes proporcionar una lista de elementos (como palabras) y luego utilizar métodos de esta clase para realizar varias operaciones, como contar la frecuencia de cada elemento, encontrar las palabras más comunes, calcular la frecuencia acumulativa, entre otros.

Resumen: **Cálculo del número de ocurrencias de cada palabra (token) dentro de un texto (sentencia).**

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

sent = 'This is an example sentence'
fdist = FreqDist()

for word in word_tokenize(sent):
  fdist[word.lower()] += 1
```

# Collocations

Las colocaciones son secuencias de palabras que tienden a aparecer juntas con más frecuencia de lo que se esperaría al azar.
Las **colocaciones** son más específicas que los **bigramas** y pueden incluir secuencias de más de dos palabras.

NLTK proporciona una función llamada **collocations()** que te permite identificar las colocaciones en un corpus de texto. Esta función utiliza medidas estadísticas como el PMI (Pointwise Mutual Information) para identificar las combinaciones de palabras más significativas.

```python
# Ejemplo de BigramCollocationFinder (bigramas)

from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

# Módulo dentro de NLTK que se utiliza para la tokenización de texto en oraciones y palabras
nltk.download('punkt')


texto = "El perro corre en el parque. El gato duerme en el sofá."
tokens = nltk.word_tokenize(texto)

finder = BigramCollocationFinder.from_words(tokens)
collocations = finder.nbest(BigramAssocMeasures.likelihood_ratio, 2)
print(collocations)
```

```python
from nltk.collocations import TrigramCollocationFinder
from nltk.metrics import TrigramAssocMeasures

texto = "El gato duerme en la cama. El perro juega en el jardín. El gato y el perro son amigos."

palabras = nltk.word_tokenize(texto)

finder = TrigramCollocationFinder.from_words(palabras)
colocations = finder.nbest(TrigramAssocMeasures.likelihood_ratio, 10)

for colocation in colocations:
    print(colocation)
```
