# spaCy I

spaCy es una biblioteca de procesamiento de lenguaje natural (NLP) en Python que se utiliza para realizar tareas como tokenización, lematización, análisis gramatical y reconocimiento de entidades.

```python
import spacy

nlp = spacy.blank("es")
```

# Doc y Tokens

Un Doc es el objeto central que spaCy utiliza para representar un texto analizado. Contiene la información procesada del texto, como los tokens individuales, sus lemas o información gramatical.

Los "tokens" en spaCy son las unidades básicas de texto después de la tokenización. La tokenización es el proceso de dividir un texto en unidades significativas, como palabras o signos de puntuación. Cada elemento individual resultante después de la tokenización se llama token. Estos tokens son las unidades fundamentales que spaCy utiliza para realizar análisis y procesamiento en el texto.

image.png

```python
doc = nlp("Hola mundo, qué tal")

for token in doc:
  print(token.text)

t = doc[1]
print(t.text)
```

# Span

Un span es una porción específica del texto, extraída de un Doc.

image.png

```python
doc = nlp("Esto es un ejemplo de texto en castellano")

span = doc[3:6]
print(span.text)
```

# Atributos léxicos de los tokens

Un objeto de tipo token tiene una serie de atributos, como son:

* El índice dentro del texto (posición)
* El texto que contiene el token
* Si es alfanumérico
* Si es un símbolo de puntuación
* Si se trata de un token numérico

```python
doc = nlp("Esto, es un ejemplo de texto en castellano 1 2 3.")

for token in doc:
  print("Índice:   ", token.i)
  print("Texto:    ", token.text)

  print("is_alpha:", token.is_alpha)
  print("is_punct:", token.is_punct)
  print("like_num:", token.like_num)
  print("--------------------------")
```

# Pipelines entrenados

Un pipeline entrenado es un conjunto de procesadores encadenados que se aplican secuencialmente a un texto durante el procesamiento.
Incluyen operaciones de tokenización, lematización, análisis de dependencias, reconocimiento de entidades, etc.

A continuación, instalamos un modelo ya preentrenado en castellano, que podemos usar para crear nuestra variable nlp.

image.png

image.png

```python
# https://spacy.io/models/es
!pip install es_core_news_sm
```

```python
nlp = spacy.load("es_core_news_sm")
```

```python
doc = nlp("Esto es otro ejemplo escrito en castellano")
for token in doc:
  print(token.text, token.pos_, token.dep_, token.head.text)
```

```python
doc = nlp("Esto es otro ejemplo escrito en castellano")
for token in doc:
  print(token.text, token.pos_, spacy.explain(token.pos_))
  print(token.text, token.dep_, spacy.explain(token.dep_))
```

```python
print(spacy.explain("LOC"))
print(spacy.explain("MISC"))
print(spacy.explain("NNP"))
print(spacy.explain("ORG"))
```

# Entidades

Una entidad o ent se refiere a una unidad significativa que representa a un concepto como nombres de personas, organizaciones, lugares, fechas o cantidades.

Cuando spaCy procesa un texto, detecta automáticamente las entidades y le asigna una etiqueta que indica su tipo concreto, por ejemplo "PER" para los nombres de personas o "ORG" para las organizaciones.

```python
doc = nlp("Esto es un texto que habla de Madrid y su relación con Federico, que trabajaba en Google")
```

```python
for ent in doc.ents:
  print(ent.text, ent.label_)
```
